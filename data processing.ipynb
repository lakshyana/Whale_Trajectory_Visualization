```python
bucket_name = 'ceti-data'
directory = "raw/"

# Get a list of all objects in the specified S3 directory
# objects = s3_client.list_objects_v2(Bucket=bucket_name, Prefix=directory) # note this limits the total objects retrieved to 1000

# Exclude the "exclude-folder" folder
exclude_prefix = directory + "hash/"

# Set the S3 client and resource
s3_client = boto3.client('s3')
s3_resource = boto3.resource('s3')

paginator = s3_client.get_paginator('list_objects_v2')
result_iterator = paginator.paginate(Bucket=bucket_name, Prefix=directory)

# Loop through the results and print the object keys
for result in result_iterator:

    # Loop through the objects in the result
    for obj in result.get('Contents', []):
        key = obj.get('Key', '')

        # Skip objects in the "exclude-folder" folder
        if key.startswith(exclude_prefix):
            continue

        print(key)
```

```python
import boto3
import os

def download_dir(client, resource, dist, local='/tmp', bucket='your_bucket'):
    paginator = client.get_paginator('list_objects')

    for result in paginator.paginate(Bucket=bucket, Delimiter='/', Prefix=dist):
        if result.get('CommonPrefixes') is not None:

            for subdir in result.get('CommonPrefixes'):
                download_dir(client, resource, subdir.get('Prefix'), local, bucket)

        for file in result.get('Contents', []):
            dest_pathname = os.path.join(local, file.get('Key'))

            if not os.path.exists(os.path.dirname(dest_pathname)):
                os.makedirs(os.path.dirname(dest_pathname))

            if not file.get('Key').endswith('/'):
                resource.meta.client.download_file(bucket, file.get('Key'), dest_pathname)

def _start():
    client = boto3.client('s3')
    resource = boto3.resource('s3')
    download_dir(client, resource, '2014', './data', bucket='whaleproject')
    download_dir(client, resource, '2016/', './data', bucket='whaleproject')
    download_dir(client, resource, '2017/', './data', bucket='whaleproject')
    download_dir(client, resource, '2018/', './data', bucket='whaleproject')
```

```python

import os
import numpy as np
import soundfile as sf

# Set the block size in frames (samples)
block_size = 10 * 44100  # 10 seconds at 44.1 kHz

for file in os.listdir('./data/2015/')[:1]:
    print(file)
    if file.endswith(".flac"):
        print(os.path.join('./data/2015/', file))
        # Open the FLAC file
        with sf.SoundFile(os.path.join('./data/2015/', file)) as f:
            # Print some information about the file
            print(f"Sample rate: {f.samplerate}")
            print(f"Number of channels: {f.channels}")
            print(f"Number of frames: {f.frames}")

            # Read the file in blocks and concatenate the audio data
            audio_data = []
            block_start = 0
            while block_start < f.frames:
                block_end = min(block_start + block_size, f.frames)
                audio_block, _ = sf.blocks(f, blocksize=block_size, start=block_start, stop=block_end)
                audio_data.append(audio_block)
                block_start = block_end
            audio_data = np.concatenate(audio_data)

            print("concatenated audio data" + str(audio_data.shape))

            # Export the audio data as a WAV file
            output_filename = os.path.splitext(file)[0] + '.wav'
            sf.write(output_filename, audio_data, f.samplerate)
            print(f"Exported audio as {output_filename}")

        print("loaded file")
```
